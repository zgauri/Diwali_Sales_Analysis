Exploratory Data Analysis

Gender:

```python
df.columns
```

```python
sns.countplot(x = 'Gender', data=df)
sns.set(rc={'figure.figsize' : (5,10)})
```

```python
ax = sns.countplot(x = 'Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

From the above graphs we can see that most buyers are females and also females have spent more as compared to men.

Age:

```python
sns.countplot(data=df, x = 'Age Group')
```

```python
sns.countplot(data=df, x = 'Age Group', hue = 'Gender')
```

```python
ax = sns.countplot(data=df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

From above graphs we can see that most of the buyers are of the age group between 26-35 years female.

State:

```python
df.replace('Andhraï¿½Pradesh', 'Andhra Pradesh', inplace=True)
df.to_csv('Diwali Sales Data.csv', index= False)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)

sns.set(rc = {'figure.figsize': (19, 5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)

sns.set(rc = {'figure.figsize': (19, 5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount')
```

From the above data we can see that unexpectedly most of the orders and total sales/amountis from Uttar Pradesh, Karnataka and then Maharastra*

Marital Status:

```python
ax = sns.countplot(data=df, x = 'Marital_Status')

sns.set(rc = {'figure.figsize': (4, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc = {'figure.figsize': (4, 5)})
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue='Gender')
```

From the above graphs we see that most of the buyers are married (women) and they have higher purchasing power.

Occupation:

```python
sns.set(rc={'figure.figsize': (20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc={'figure.figsize': (20,5)})
sns.barplot(data= sales_state, x = 'Occupation', y = 'Amount')
```

From the above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sectors.

Product Category:

```python
df.columns
```

```python
sns.set(rc={'figure.figsize' :(27, 5)})
ax = sns.countplot(data= df, x = 'Product_Category')
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc={'figure.figsize': (34,5)})
sns.barplot(data= sales_state, x = 'Product_Category', y = 'Amount')
```

From the above graphs we can see that most of the products are from food, clothing and electronics category.

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False)

sns.set(rc={'figure.figsize' : (20, 5)})
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders')
```

```python
fig1, ax1 = plt.subplots(figsize=(12, 7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
